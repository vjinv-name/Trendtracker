### ë¹„êµ ì—°ì‚°ì (ê²°ê³¼: True / False)
| ì—°ì‚°ì | ì„¤ëª… | ì˜ˆì‹œ | ê²°ê³¼ |
|---------|------|------|------|
| `==` | ê°™ë‹¤ | `5 == 5` | `True` |
| `!=` | ë‹¤ë¥´ë‹¤ | `5 != 3` | `True` |

````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : ì‚°ìˆ  ì—°ì‚°ì
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

a = 17
b = 5

print(f"{a} + {b} = {a + b}")    # ë§ì…ˆ: 22
print(f"{a} - {b} = {a - b}")    # ëº„ì…ˆ: 12
print(f"{a} * {b} = {a * b}")    # ê³±ì…ˆ: 85
print(f"{a} / {b} = {a / b}")    # ë‚˜ëˆ—ì…ˆ: 3.4 (í•­ìƒ float ë°˜í™˜)
print(f"{a} // {b} = {a // b}")  # ëª¹: 3
print(f"{a} % {b} = {a % b}")    # ë‚˜ë¨¸ì§€: 2
print(f"{a} ** {b} = {a ** b}")  # ê±°ë“­ì œê³±: 1419857
````

````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : ë¹„êµ ì—°ì‚°ì
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

x = 10
y = 20

print(f"{x} == {y} â†’ {x == y}")   # False
print(f"{x} != {y} â†’ {x != y}")   # True
print(f"{x} < {y}  â†’ {x < y}")    # True
print(f"{x} > {y}  â†’ {x > y}")    # False
print(f"{x} <= {y} â†’ {x <= y}")   # True
print(f"{x} >= {y} â†’ {x >= y}")   # False

# ë¹„êµ ì—°ì‚°ìì˜ ê²°ê³¼ëŠ” í•­ìƒ bool(True/False)
result = (x < y)
print(f"\nê²°ê³¼ì˜ ìë£Œí˜•: {type(result)}")  # <class 'bool'>
````

````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : ë…¼ë¦¬ ì—°ì‚°ì
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

age = 25
has_id = True

# and: ë‘˜ ë‹¤ Trueì—¬ì•¼ True
print(f"age >= 19 and has_id â†’ {age >= 19 and has_id}")  # True

# or: í•˜ë‚˜ë§Œ Trueì´ì–´ë„ True
print(f"age < 19 or has_id  â†’ {age < 19 or has_id}")    # True

# not: ë°˜ëŒ€ë¡œ
print(f"not has_id          â†’ {not has_id}")              # False

# ë³µí•© ì¡°ê±´ ì˜ˆì‹œ
score = 85
attendance = 90
print(f"\nì„±ì : {score}, ì¶œì„ë¥ : {attendance}%")
print(f"í•©ê²© ì—¬ë¶€ (80ì  ì´ìƒ AND ì¶œì„ 75% ì´ìƒ): {score >= 80 and attendance >= 75}")
````

````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : í• ë‹¹ ì—°ì‚°ì
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

count = 10
print(f"ì´ˆê¸°ê°’: {count}")

count += 5   # count = count + 5
print(f"+= 5 í›„: {count}")   # 15

count -= 3   # count = count - 3
print(f"-= 3 í›„: {count}")   # 12

count *= 2   # count = count * 2
print(f"*= 2 í›„: {count}")   # 24

count /= 4   # count = count / 4
print(f"/= 4 í›„: {count}")   # 6.0 (ë‚˜ëˆ—ì…ˆì€ í•­ìƒ float)
````
