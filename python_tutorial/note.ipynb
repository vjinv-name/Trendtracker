```python

#실습 2-1-1
score = int(input("점수를 입력하세요: "))
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"


print(f"{score}점은 {grade}학점 입니다.")
```

```python
#실습 2-1-2
number = int(input("숫자를 입력하세요: "))

if number % 2 == 0:
    result = ("짝수입니다.")
elif number % 2 == 1:
    result = ("홀수입니다.")
elif number == 0:
    result = ("영(제로)입니다")

print(f"입력: {number} -> {number}은(는) {result}")
```
