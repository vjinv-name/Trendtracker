## 3-1. 리스트

### 📖 이론: 리스트란?

**리스트(list)** 는 파이썬에서 가장 많이 사용되는 자료구조입니다.  
여러 개의 값을 **순서대로** 저장하며, 저장된 값을 자유롭게 **추가, 수정, 삭제**할 수 있습니다.

#### 핵심 특징
| 특징 | 설명 |
|------|------|
| **순서 있음** (ordered) | 인덱스(0부터 시작)로 각 요소에 접근할 수 있습니다 |
| **변경 가능** (mutable) | 생성 후에도 요소를 추가·수정·삭제할 수 있습니다 |
| **중복 허용** | 같은 값을 여러 번 저장할 수 있습니다 |
| **다양한 타입 혼합** | 하나의 리스트에 정수, 문자열, 불리언 등을 섞어 담을 수 있습니다 |

#### 주요 메서드 요약
| 메서드 | 설명 | 예시 |
|--------|------|------|
| `append(x)` | 맨 끝에 요소 추가 | `lst.append(5)` |
| `insert(i, x)` | i 위치에 요소 삽입 | `lst.insert(0, 'a')` |
| `remove(x)` | 값이 x인 첫 번째 요소 제거 | `lst.remove(3)` |
| `pop(i)` | i 위치 요소를 꺼내고 반환 (기본: 마지막) | `lst.pop()` |
| `del lst[i]` | i 위치 요소 삭제 (반환 없음) | `del lst[2]` |
| `sort()` | 리스트 자체를 정렬 (원본 변경) | `lst.sort()` |
| `sorted(lst)` | 정렬된 **새 리스트** 반환 (원본 유지) | `new = sorted(lst)` |
| `reverse()` | 리스트 순서 뒤집기 | `lst.reverse()` |
| `len(lst)` | 리스트 길이(요소 개수) | `len(lst)` |
| `x in lst` | x가 리스트에 있는지 확인 | `3 in lst` |

````python
# ──────────────────────────────────────
# 📖 이론: 리스트 생성 방법
# ──────────────────────────────────────

# 1) 대괄호 []로 생성
fruits = ["사과", "바나나", "딸기", "포도"]
print("과일 리스트:", fruits)

# 2) 빈 리스트 생성
empty1 = []
empty2 = list()
print("빈 리스트 1:", empty1)
print("빈 리스트 2:", empty2)

# 3) list() 함수로 다른 자료형을 리스트로 변환
chars = list("Python")
print("문자열 → 리스트:", chars)

nums_from_range = list(range(1, 6))
print("range → 리스트:", nums_from_range)

# 4) 다양한 타입 혼합
mixed = [1, "hello", 3.14, True, None]
print("혼합 리스트:", mixed)
````

````python
# ──────────────────────────────────────
# 📖 이론: 인덱싱과 슬라이싱
# ──────────────────────────────────────

colors = ["빨강", "주황", "노랑", "초록", "파랑", "남색", "보라"]

# --- 인덱싱 ---
# 양수 인덱스: 앞에서부터 0, 1, 2, ...
print("첫 번째 색:", colors[0])    # 빨강
print("세 번째 색:", colors[2])    # 노랑

# 음수 인덱스: 뒤에서부터 -1, -2, ...
print("마지막 색:", colors[-1])    # 보라
print("뒤에서 두 번째:", colors[-2])  # 남색

print()

# --- 슬라이싱 [start:end:step] ---
# start 포함, end 미포함
print("처음 3개:", colors[0:3])     # ['빨강', '주황', '노랑']
print("처음 3개:", colors[:3])      # 위와 동일 (start 생략 = 0)
print("3번째부터:", colors[2:])     # ['노랑', '초록', '파랑', '남색', '보라']
print("2개씩 건너뛰기:", colors[::2])  # ['빨강', '노랑', '파랑', '보라']
print("역순:", colors[::-1])        # ['보라', '남색', '파랑', '초록', '노랑', '주황', '빨강']
````

````python
# ──────────────────────────────────────
# 📖 이론: 리스트 CRUD (추가/수정/삭제)
# ──────────────────────────────────────

# 리스트 준비
animals = ["고양이", "강아지", "토끼"]
print("원본:", animals)

# --- Create (추가) ---
# append: 맨 뒤에 추가
animals.append("햄스터")
print("append 후:", animals)

# insert: 원하는 위치에 삽입
animals.insert(1, "앵무새")  # 인덱스 1 위치에 삽입
print("insert 후:", animals)

print()

# --- Update (수정) ---
# 인덱스로 직접 수정
animals[0] = "페르시안 고양이"
print("수정 후:", animals)

print()

# --- Delete (삭제) ---
# remove: 값으로 삭제 (첫 번째 일치 항목)
animals.remove("앵무새")
print("remove 후:", animals)

# pop: 인덱스로 삭제하고 삭제된 값 반환
removed = animals.pop()    # 기본: 마지막 요소
print(f"pop으로 꺼낸 값: {removed}")
print("pop 후:", animals)

# del: 인덱스로 삭제 (반환 없음)
del animals[0]
print("del 후:", animals)
````

````python
# ──────────────────────────────────────
# 📖 이론: 정렬 (sort vs sorted)
# ──────────────────────────────────────

scores = [85, 92, 78, 95, 88, 70]
print("원본:", scores)

# sorted(): 새로운 정렬된 리스트를 반환 (원본 유지!)
sorted_scores = sorted(scores)
print("sorted() 결과:", sorted_scores)
print("원본 유지 확인:", scores)  # 원본은 그대로

# 내림차순 정렬
desc_scores = sorted(scores, reverse=True)
print("내림차순:", desc_scores)

print()

# sort(): 원본 리스트 자체를 정렬 (반환값 None)
scores.sort()
print("sort() 후 원본:", scores)  # 원본이 변경됨!

# reverse(): 리스트 순서 뒤집기
scores.reverse()
print("reverse() 후:", scores)

print()

# ⚠️ 핵심 차이
# sort()  → 원본을 변경, 반환값 None
# sorted() → 원본 유지, 새 리스트 반환
print("--- sort vs sorted 핵심 차이 ---")
nums = [3, 1, 4, 1, 5]
result = nums.sort()   # result는 None!
print(f"sort() 반환값: {result}")
print(f"nums 변경됨: {nums}")
````

````python
# ──────────────────────────────────────
# 📖 이론: len(), in 연산자, 중첩 리스트
# ──────────────────────────────────────

# --- len(): 리스트 길이 ---
items = ["연필", "지우개", "자", "가위", "풀"]
print(f"아이템 개수: {len(items)}개")

print()

# --- in 연산자: 포함 여부 확인 ---
print("'연필' 있나?", "연필" in items)       # True
print("'볼펜' 있나?", "볼펜" in items)       # False
print("'볼펜' 없나?", "볼펜" not in items)   # True

# 조건문과 함께 자주 사용
search = "자"
if search in items:
    print(f"'{search}'이(가) 목록에 있습니다!")
else:
    print(f"'{search}'이(가) 목록에 없습니다.")

print()

# --- 중첩 리스트 (리스트 안의 리스트) ---
# 2차원 데이터(행렬, 표 등)를 표현할 때 유용
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("전체 행렬:", matrix)
print("첫 번째 행:", matrix[0])        # [1, 2, 3]
print("두 번째 행, 세 번째 열:", matrix[1][2])  # 6

# 중첩 리스트 순회
print("\n행렬 출력:")
for row in matrix:
    for val in row:
        print(f"{val:3d}", end="")
    print()  # 줄바꿈
````

## 3-2. 튜플과 집합

### 📖 이론: 튜플(tuple)이란?

**튜플(tuple)** 은 리스트와 비슷하지만, 한 번 만들면 **변경할 수 없는(immutable)** 자료구조입니다.  
변경이 불필요한 데이터를 안전하게 보관할 때 사용합니다.

#### 튜플 vs 리스트
| 구분 | 리스트 `[]` | 튜플 `()` |
|------|-------------|----------|
| 변경 가능 여부 | 가능 (mutable) | 불가능 (immutable) |
| 기호 | 대괄호 `[]` | 소괄호 `()` |
| 속도 | 상대적으로 느림 | 상대적으로 빠름 |
| 용도 | 수정이 필요한 데이터 | 수정이 불필요한 고정 데이터 |

#### 튜플의 활용
- 좌표 `(x, y)`, 색상 `(R, G, B)` 등 고정된 묶음 데이터
- 함수에서 여러 값을 동시에 반환할 때
- 딕셔너리의 키로 사용 (리스트는 불가능)

---

### 📖 이론: 집합(set)이란?

**집합(set)** 은 수학의 집합과 동일한 개념으로, **중복을 허용하지 않고** **순서가 없는** 자료구조입니다.

#### 핵심 특징
| 특징 | 설명 |
|------|------|
| **중복 없음** | 같은 값을 여러 번 넣어도 하나만 유지 |
| **순서 없음** | 인덱싱으로 접근 불가 |
| **집합 연산 지원** | 합집합, 교집합, 차집합 등 |

#### 집합 연산
| 연산 | 연산자 | 메서드 | 설명 |
|------|--------|--------|------|
| 합집합 | `A \| B` | `A.union(B)` | A 또는 B에 있는 모든 요소 |
| 교집합 | `A & B` | `A.intersection(B)` | A와 B 모두에 있는 요소 |
| 차집합 | `A - B` | `A.difference(B)` | A에만 있는 요소 |
| 대칭차집합 | `A ^ B` | `A.symmetric_difference(B)` | A 또는 B 중 하나에만 있는 요소 |

````python
# ──────────────────────────────────────
# 📖 이론: 튜플 기본 사용
# ──────────────────────────────────────

# 1) 튜플 생성
point = (3, 5)
colors = ("빨강", "초록", "파랑")
single = (42,)  # 요소가 하나인 튜플 → 쉼표 필수!
not_tuple = (42) # 이건 그냥 정수 42

print(f"좌표: {point}, 타입: {type(point)}")
print(f"색상: {colors}")
print(f"단일 요소 튜플: {single}, 타입: {type(single)}")
print(f"괄호만 사용: {not_tuple}, 타입: {type(not_tuple)}")

print()

# 2) 인덱싱 (리스트와 동일)
print("첫 번째 색:", colors[0])
print("마지막 색:", colors[-1])

print()

# 3) 불변성 확인 - 수정 시도하면 에러 발생
try:
    colors[0] = "노랑"  # TypeError!
except TypeError as e:
    print(f"튜플 수정 시도 → 에러: {e}")
````

````python
# ──────────────────────────────────────
# 📖 이론: 패킹, 언패킹, 다중 반환
# ──────────────────────────────────────

# --- 패킹(packing): 여러 값을 튜플로 묶기 ---
person = "홍길동", 25, "서울"  # 괄호 없이도 튜플 생성 가능
print(f"패킹: {person}, 타입: {type(person)}")

# --- 언패킹(unpacking): 튜플을 여러 변수에 풀기 ---
name, age, city = person
print(f"이름: {name}, 나이: {age}, 도시: {city}")

# 일부만 받고 나머지는 *로 묶기
first, *rest = (1, 2, 3, 4, 5)
print(f"첫 번째: {first}, 나머지: {rest}")

print()

# --- 함수에서 다중 반환값 ---
def min_max(numbers):
    """리스트의 최솟값과 최댓값을 동시에 반환"""
    return min(numbers), max(numbers)  # 튜플로 반환

data = [45, 78, 23, 91, 56]
minimum, maximum = min_max(data)  # 언패킹으로 받기
print(f"최솟값: {minimum}, 최댓값: {maximum}")

# 그대로 튜플로 받기
result = min_max(data)
print(f"결과 튜플: {result}, 타입: {type(result)}")
````

````python
# ──────────────────────────────────────
# 📖 이론: 집합(set) 생성과 기본 연산
# ──────────────────────────────────────

# 1) 집합 생성
fruits = {"사과", "바나나", "딸기", "사과"}  # 중복은 자동 제거
print("집합:", fruits)  # 순서는 보장되지 않음

# 빈 집합은 반드시 set()으로! ({}는 빈 딕셔너리)
empty_set = set()
empty_dict = {}
print(f"set() 타입: {type(empty_set)}")
print(f"{{}} 타입: {type(empty_dict)}")

print()

# 2) 요소 추가/삭제
animals = {"고양이", "강아지"}
animals.add("토끼")       # 추가
animals.add("고양이")     # 이미 있으므로 변화 없음
print("add 후:", animals)

animals.remove("강아지")  # 삭제 (없으면 에러)
print("remove 후:", animals)

animals.discard("호랑이") # 삭제 (없어도 에러 안남)
print("discard 후:", animals)
````

````python
# ──────────────────────────────────────
# 📖 이론: 집합 연산 시연
# ──────────────────────────────────────

# 두 동아리의 회원 명단
music_club = {"지민", "수현", "태영", "하은", "도윤"}
art_club = {"수현", "채린", "태영", "민서"}

print("음악 동아리:", music_club)
print("미술 동아리:", art_club)
print()

# 합집합: 둘 중 하나에라도 속한 학생
print("합집합 (|)  :", music_club | art_club)

# 교집합: 두 동아리 모두에 속한 학생
print("교집합 (&)  :", music_club & art_club)

# 차집합: 음악만, 미술만
print("음악만 (-)  :", music_club - art_club)
print("미술만 (-)  :", art_club - music_club)

# 대칭차집합: 하나의 동아리에만 속한 학생
print("대칭차집합 (^):", music_club ^ art_club)

print()

# --- 리스트 중복 제거 활용 ---
scores = [85, 92, 78, 85, 92, 100, 78]
unique_scores = list(set(scores))  # set으로 중복 제거 후 다시 리스트로
print(f"원본: {scores}")
print(f"중복 제거: {unique_scores}")
````

## 3-3. 딕셔너리

### 📖 이론: 딕셔너리(dictionary)란?

**딕셔너리(dict)** 는 **키(key)** 와 **값(value)** 의 쌍으로 데이터를 저장하는 자료구조입니다.  
실생활의 사전(dictionary)처럼, 단어(키)로 뜻(값)을 찾는 방식과 같습니다.

#### 핵심 특징
| 특징 | 설명 |
|------|------|
| **키-값 쌍** | 각 데이터는 `key: value` 형태 |
| **키 중복 불가** | 같은 키가 두 번 나오면 마지막 값으로 덮어씀 |
| **키는 불변 타입** | 문자열, 숫자, 튜플 가능 / 리스트, 집합 불가 |
| **변경 가능** | 생성 후 추가, 수정, 삭제 가능 |
| **순서 유지** | Python 3.7+부터 삽입 순서 유지 |

#### 주요 메서드
| 메서드 | 설명 |
|--------|------|
| `d[key]` | 키로 값 접근 (없으면 KeyError) |
| `d.get(key, default)` | 키로 값 접근 (없으면 default 반환, 기본 None) |
| `d[key] = value` | 추가 또는 수정 |
| `del d[key]` | 키-값 삭제 |
| `d.pop(key)` | 키-값 삭제하고 값 반환 |
| `d.keys()` | 모든 키 반환 |
| `d.values()` | 모든 값 반환 |
| `d.items()` | 모든 (키, 값) 쌍 반환 |
| `key in d` | 키 존재 여부 확인 |

````python
# ──────────────────────────────────────
# 📖 이론: 딕셔너리 생성과 접근
# ──────────────────────────────────────

# 1) 중괄호 {}로 생성
student = {
    "이름": "김지수",
    "나이": 20,
    "전공": "컴퓨터공학",
    "학년": 2
}
print("학생 정보:", student)

# 2) dict() 함수로 생성
config = dict(host="localhost", port=8080, debug=True)
print("설정:", config)

print()

# 3) 값 접근
print("이름:", student["이름"])      # 키로 접근
print("전공:", student["전공"])

# 존재하지 않는 키 → KeyError
try:
    print(student["주소"])
except KeyError as e:
    print(f"KeyError 발생: {e}")
````

````python
# ──────────────────────────────────────
# 📖 이론: 딕셔너리 CRUD
# ──────────────────────────────────────

book = {
    "제목": "파이썬 입문",
    "저자": "홍길동",
    "가격": 25000
}
print("원본:", book)

# --- Create (추가) ---
book["출판사"] = "한빛미디어"   # 새 키-값 추가
book["페이지"] = 350
print("추가 후:", book)

# --- Update (수정) ---
book["가격"] = 28000           # 기존 키에 새 값 대입 → 수정
print("수정 후:", book)

# --- Delete (삭제) ---
# del: 삭제만
del book["페이지"]
print("del 후:", book)

# pop: 삭제하고 값 반환
publisher = book.pop("출판사")
print(f"pop으로 꺼낸 값: {publisher}")
print("pop 후:", book)

# pop에 기본값 지정 (키가 없을 때 에러 방지)
isbn = book.pop("ISBN", "없음")
print(f"ISBN: {isbn}")
````

````python
# ──────────────────────────────────────
# 📖 이론: 딕셔너리 순회
# ──────────────────────────────────────

scores = {
    "국어": 90,
    "영어": 85,
    "수학": 92,
    "과학": 88
}

# --- keys(): 키만 순회 ---
print("=== 과목 목록 ===")
for subject in scores.keys():
    print(f"  - {subject}")

print()

# --- values(): 값만 순회 ---
print("=== 점수 목록 ===")
for score in scores.values():
    print(f"  - {score}점")

print()

# --- items(): 키-값 쌍 순회 (가장 많이 사용!) ---
print("=== 과목별 성적 ===")
for subject, score in scores.items():
    grade = "우수" if score >= 90 else "보통"
    print(f"  {subject}: {score}점 ({grade})")

print()

# 참고: 딕셔너리를 그냥 순회하면 키가 나옴
print("기본 순회(키):")
for key in scores:
    print(f"  {key} → {scores[key]}")
````

````python
# ──────────────────────────────────────
# 📖 이론: get() vs [] 차이
# ──────────────────────────────────────

user = {"이름": "박민수", "이메일": "minsu@example.com"}

# [] 방식: 키가 없으면 KeyError
print("이름:", user["이름"])
# print(user["전화"])  # KeyError!

# get() 방식: 키가 없으면 None 또는 기본값 반환
print("전화:", user.get("전화"))           # None
print("전화:", user.get("전화", "미등록"))  # "미등록"
print("이름:", user.get("이름", "없음"))    # "박민수" (있으면 해당 값)

print()

# 실전 활용: 안전한 접근
settings = {"theme": "dark", "language": "ko"}

# get()으로 기본값 설정 → 에러 없이 안전하게 처리
font_size = settings.get("font_size", 14)
print(f"폰트 크기: {font_size}")

# in으로 먼저 확인하는 패턴
if "theme" in settings:
    print(f"테마: {settings['theme']}")
else:
    print("테마 설정 없음")
````

````python
# ──────────────────────────────────────
# 📖 이론: 중첩 딕셔너리
# ──────────────────────────────────────

# 딕셔너리 안에 딕셔너리를 넣어 복잡한 구조 표현
students = {
    "S001": {
        "이름": "김지수",
        "성적": {"국어": 90, "영어": 85, "수학": 92}
    },
    "S002": {
        "이름": "이태영",
        "성적": {"국어": 78, "영어": 92, "수학": 88}
    },
    "S003": {
        "이름": "박수현",
        "성적": {"국어": 95, "영어": 80, "수학": 85}
    }
}

# 중첩 접근
print("S001 이름:", students["S001"]["이름"])
print("S001 수학:", students["S001"]["성적"]["수학"])

print()

# 전체 학생 정보 출력
for student_id, info in students.items():
    name = info["이름"]
    scores = info["성적"]
    avg = sum(scores.values()) / len(scores)
    print(f"[{student_id}] {name}: 평균 {avg:.1f}점")
````

````python
# ──────────────────────────────────────
# 📖 이론: 딕셔너리 컴프리헨션
# ──────────────────────────────────────

# 기본 형태: {key: value for 변수 in 반복가능객체}

# 1) 숫자의 제곱 딕셔너리
squares = {n: n**2 for n in range(1, 6)}
print("제곱:", squares)
# {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# 2) 조건 포함
even_squares = {n: n**2 for n in range(1, 11) if n % 2 == 0}
print("짝수 제곱:", even_squares)

# 3) 리스트 → 딕셔너리 변환
fruits = ["사과", "바나나", "딸기"]
fruit_len = {fruit: len(fruit) for fruit in fruits}
print("과일 글자수:", fruit_len)

# 4) 두 리스트를 딕셔너리로 합치기 (zip 활용)
keys = ["이름", "나이", "도시"]
values = ["홍길동", 25, "서울"]
person = {k: v for k, v in zip(keys, values)}
print("zip 활용:", person)

# 참고: dict(zip(...))으로 더 간단하게
person2 = dict(zip(keys, values))
print("dict+zip:", person2)
````

## 📝 Step 3 요약

### 자료구조 비교

| 자료구조 | 기호 | 순서 | 변경 | 중복 | 주요 용도 |
|----------|------|------|------|------|----------|
| **리스트** | `[]` | O | O | O | 순서가 있는 데이터 모음 |
| **튜플** | `()` | O | X | O | 변경 불필요한 고정 데이터 |
| **집합** | `{}` | X | O | X | 중복 제거, 집합 연산 |
| **딕셔너리** | `{k:v}` | O* | O | 키 X | 키-값 매핑 데이터 |

\* Python 3.7+부터 삽입 순서 유지

### 핵심 정리
- **리스트**: 가장 범용적인 자료구조. `append`, `remove`, `sort` 등으로 자유롭게 조작
- **튜플**: 불변 데이터에 사용. 함수의 다중 반환값, 좌표 등에 활용
- **집합**: 중복 제거와 집합 연산(`|`, `&`, `-`, `^`)에 특화
- **딕셔너리**: 키-값 쌍으로 데이터 관리. `get()`으로 안전하게 접근
