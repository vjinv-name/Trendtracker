## ğŸ“– ì´ë¡ 

### ë³€ìˆ˜ ìŠ¤ì½”í”„
- **ì§€ì—­ ë³€ìˆ˜**: í•¨ìˆ˜ ì•ˆì—ì„œ ì •ì˜, í•¨ìˆ˜ ì•ˆì—ì„œë§Œ ì‚¬ìš© ê°€ëŠ¥
- **ì „ì—­ ë³€ìˆ˜**: í•¨ìˆ˜ ë°”ê¹¥ì—ì„œ ì •ì˜, ì–´ë””ì„œë“  ì½ê¸° ê°€ëŠ¥
- `global` í‚¤ì›Œë“œ: í•¨ìˆ˜ ì•ˆì—ì„œ ì „ì—­ ë³€ìˆ˜ ìˆ˜ì • ì‹œ ì‚¬ìš©

### lambda í•¨ìˆ˜
í•œ ì¤„ì§œë¦¬ ìµëª… í•¨ìˆ˜ì…ë‹ˆë‹¤.
```python
lambda ë§¤ê°œë³€ìˆ˜: í‘œí˜„ì‹
```

### ìœ ìš©í•œ ë‚´ì¥ í•¨ìˆ˜
| í•¨ìˆ˜ | ì„¤ëª… |
|------|------|
| `map(func, iterable)` | ëª¨ë“  ìš”ì†Œì— í•¨ìˆ˜ ì ìš© |
| `filter(func, iterable)` | ì¡°ê±´ì´ Trueì¸ ìš”ì†Œë§Œ |
| `sorted(iterable, key=func)` | key ê¸°ì¤€ ì •ë ¬ |

````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : ì§€ì—­ ë³€ìˆ˜ vs ì „ì—­ ë³€ìˆ˜
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

x = 10  # ì „ì—­ ë³€ìˆ˜

def my_func():
    x = 20  # ì§€ì—­ ë³€ìˆ˜ (ì „ì—­ xì™€ ë‹¤ë¥¸ ë³€ìˆ˜)
    print(f"í•¨ìˆ˜ ë‚´ë¶€ x = {x}")

my_func()
print(f"í•¨ìˆ˜ ì™¸ë¶€ x = {x}")  # ì „ì—­ xëŠ” ë³€í•˜ì§€ ì•ŠìŒ

# global í‚¤ì›Œë“œ
count = 0

def increment():
    global count
    count += 1

increment()
increment()
print(f"count = {count}")  # 2
````

````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : lambda, map, filter, sorted
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

# lambda ê¸°ë³¸
square = lambda x: x ** 2
print(f"5ì˜ ì œê³±: {square(5)}")

# map: ëª¨ë“  ìš”ì†Œì— í•¨ìˆ˜ ì ìš©
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print(f"ì œê³±: {squared}")

# filter: ì¡°ê±´ì— ë§ëŠ” ìš”ì†Œë§Œ
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"ì§ìˆ˜: {evens}")

# sorted with key
words = ["banana", "apple", "cherry", "date"]
by_length = sorted(words, key=lambda w: len(w))
print(f"ê¸¸ì´ìˆœ ì •ë ¬: {by_length}")

# ë”•ì…”ë„ˆë¦¬ ë¦¬ìŠ¤íŠ¸ ì •ë ¬
students = [
    {"name": "í™ê¸¸ë™", "score": 85},
    {"name": "ê¹€ì² ìˆ˜", "score": 92},
    {"name": "ì´ì˜í¬", "score": 78},
]
by_score = sorted(students, key=lambda s: s["score"], reverse=True)
for s in by_score:
    print(f"  {s["name"]}: {s["score"]}ì ")
````

````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ğŸ“– ì´ë¡ : random ëª¨ë“ˆ
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

import random

# randint: ë²”ìœ„ ë‚´ ëœë¤ ì •ìˆ˜
print(f"ì£¼ì‚¬ìœ„: {random.randint(1, 6)}")

# choice: ë¦¬ìŠ¤íŠ¸ì—ì„œ ëœë¤ ì„ íƒ
menu = ["ì§œì¥ë©´", "ì§¬ë½•", "ë³¶ìŒë°¥", "íƒ•ìˆ˜ìœ¡"]
print(f"ì˜¤ëŠ˜ì˜ ë©”ë‰´: {random.choice(menu)}")

# sample: ì¤‘ë³µ ì—†ì´ ì—¬ëŸ¬ ê°œ ì„ íƒ
lotto = random.sample(range(1, 46), 6)
print(f"ë¡œë˜ ë²ˆí˜¸: {sorted(lotto)}")

# shuffle: ë¦¬ìŠ¤íŠ¸ ì„ê¸°
cards = list(range(1, 11))
random.shuffle(cards)
print(f"ì„ì€ ì¹´ë“œ: {cards}")
````
